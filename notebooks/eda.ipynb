# 🛒 Retail Sales Forecasting — Exploratory Data Analysis
**ML Project**

This notebook explores the retail sales dataset before model training.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 110

df = pd.read_csv('../dataset/sales_data.csv', parse_dates=['date'])
print(f'Dataset shape: {df.shape}')
df.head()
```

```python
# Basic info
df.info()
print('\nMissing values:')
print(df.isnull().sum())
```

```python
# Statistical summary
df.describe()
```

```python
# Sales distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

df['sales'].hist(bins=40, ax=axes[0], color='steelblue', edgecolor='white')
axes[0].set_title('Sales Distribution', fontsize=13)
axes[0].set_xlabel('Sales')

df.boxplot(column='sales', by='store_id', ax=axes[1])
axes[1].set_title('Sales by Store')

plt.tight_layout()
plt.show()
```

```python
# Monthly sales trend
df['month'] = df['date'].dt.to_period('M')
monthly = df.groupby('month')['sales'].mean().reset_index()
monthly['month'] = monthly['month'].astype(str)

plt.figure(figsize=(14, 4))
plt.plot(monthly['month'], monthly['sales'], marker='o', color='steelblue', linewidth=2)
plt.xticks(rotation=45)
plt.title('Average Monthly Sales Over Time', fontsize=13)
plt.ylabel('Avg Sales')
plt.tight_layout()
plt.show()
```

```python
# Sales by day of week
df['day_of_week'] = df['date'].dt.day_name()
dow_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
dow_sales = df.groupby('day_of_week')['sales'].mean().reindex(dow_order)

plt.figure(figsize=(9, 4))
dow_sales.plot(kind='bar', color='mediumpurple', edgecolor='white')
plt.title('Average Sales by Day of Week', fontsize=13)
plt.ylabel('Avg Sales')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# Promotion impact
promo_impact = df.groupby('promotion')['sales'].mean()
promo_impact.index = ['No Promotion', 'Promotion']

plt.figure(figsize=(6, 4))
promo_impact.plot(kind='bar', color=['steelblue', 'tomato'], edgecolor='white')
plt.title('Average Sales: Promotion vs No Promotion', fontsize=13)
plt.ylabel('Avg Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

lift = (promo_impact['Promotion'] - promo_impact['No Promotion']) / promo_impact['No Promotion'] * 100
print(f'Promotion Sales Lift: +{lift:.1f}%')
```

```python
# Correlation heatmap
df_num = df[['sales','promotion','holiday']].copy()
df_num['month_num'] = df['date'].dt.month
df_num['day_of_week_num'] = df['date'].dt.dayofweek

plt.figure(figsize=(7, 5))
sns.heatmap(df_num.corr(), annot=True, fmt='.3f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# Product-level sales comparison
prod_sales = df.groupby('product_id')['sales'].mean()

plt.figure(figsize=(7, 4))
prod_sales.plot(kind='bar', color='teal', edgecolor='white')
plt.title('Average Sales by Product', fontsize=13)
plt.ylabel('Avg Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
